```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
θ = np.linspace(0,2*np.pi*np.pi,100)
r = 1-10*np.sin(θ)-10*np.cos(θ)

x = r*r*np.sin(θ)
y = r*r*np.cos(θ)

ax = plt.gca()
ax.set_aspect("equal")

plt.plot(x,y)
```

```python
δ = np.pi/4

A = np.array([[np.cos(δ),-np.sin(δ)],
              [np.sin(δ),np.cos(δ)]])

xy  = np.array([x,y])
Axy = A@xy

ax = plt.gca()
ax.set_aspect("equal")

plt.plot(Axy[0],Axy[1])
```
