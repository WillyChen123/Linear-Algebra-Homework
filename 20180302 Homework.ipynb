```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline 

x = np.linspace(0,10,100)
```

```python
y1 = 5*x - 10
y2 = 5*x - 3
plt.plot(x,y1,x,y2)
```

```python
y1 = 5*x - 1
y2 = 3*x + 4
plt.plot(x,y1,x,y2)
```

```python
y1 = 5*x - 1
y2 = (10*x - 2)/2
plt.plot(x,y1,x,y2)
```
