```python
%matplotlib inline

import numpy as np
import sympy as sp
import random
import math
import matplotlib.pyplot as plt
```

```python
# Return the result of the function
def func(t):    
    # return f(x) result
    # Change target into sympy symbol
    x = sp.Symbol('x')
    target = -2*x**5 - 3*x**4 + 1520*x**3 - 6530*x**2
    return target.subs(x,t)

# Derivatives
def df(t):
    x = sp.Symbol('x')
    target = -2*x**5 - 3*x**4 + 1520*x**3 - 6530*x**2
    diff = sp.diff(target,x)
    return diff.subs(x,t)

# Gradient descent
# Start from (x_start), run (epochs) times, learning rate (lr)
def gd(x_start,epochs,lr):
    result = np.zeros(epochs+1)
    x = x_start
    result[0] = x
    for i in range(epochs):
        dx = df(x)
#         print("dx : ",dx)
        # The next times x change (u) learning rage (lr)
        u = - dx * lr
        x += u
#         print("x :",x)
        result[i+1] = x
    return result    
```

```python
# The minimum of function
minimum = func(-30)
x_min = 0
times = 10

for i in range(times):
    x_start = random.randint(-30,30)
    # Default times
    epochs = 100
    # Learning rate
    lr = 0.000001
    # Calculate gradient descent
    w = gd(x_start,epochs,lr)
    # Check if w is the minimum
    if math.isinf(func(w[-1])) or math.isnan(func(w[-1])):
        continue
    
    print(w)
    print(w[-1],minimum)
    
    if func(w[-1]) < minimum:
        x_min = w[-1]
        minimum = func(w[-1])
        
    # Show the picture
    x = np.linspace(-30,30,2000)
    y = -2*x**5 - 3*x**4 + 1520*x**3 - 6530*x**2
    plt.plot(x,y,'b')
    w2 = np.zeros(epochs+1)
    for i in range(epochs+1):
        w2[i] = func(w[i])
    plt.plot(w,w2,c='r',label='lr={}'.format(lr))
    plt.scatter(w,w2,c='r')
    plt.legend()
    plt.show()
    
print("When x =",x_min,"has the minimum value",minimum)
    
```
